## Requirement

    1. Open any image;
    2. Use Sobel to find the horizontal and vertical gradients;
    3. Create an edge mask using Canny (255 for edges 0 for not);
    4. On the edges, show the magnitude of the gradients.

    Advanced Task:
    1. Devise a way to visualise the direction of the gradients on the edges.

```python
# Import libraries

%matplotlib inline

import cv2 # The OpenCV library; install using `pip install opencv-contrib-python`
import numpy as np # Helpful when working with arrays; install using `pip install numpy`
from matplotlib import pyplot as plt # Good for graphing; install using `pip install matplotlib`
from matplotlib import image as image
import easygui # An easy-to-use file-picker; pip install easygui
```

```python
# 1. Open any image;

I = cv2.imread(r'res\ball.jpg')

G = cv2.cvtColor(src = I, code = cv2.COLOR_BGR2GRAY)

# Sobel derivatives
ddepth = cv2.CV_16S #CV_16S - 16-bit signed integers

# Calculate horizontal changes
grad_x = cv2.Sobel(G, ddepth, 1, 0, ksize=3, borderType=cv2.BORDER_DEFAULT)

# Calculate vertical changes
grad_y = cv2.Sobel(G, ddepth, 0, 1, ksize=3, borderType=cv2.BORDER_DEFAULT)

fig = plt.figure()

fig.add_subplot(121)
plt.imshow(grad_x, cmap = 'gray')

fig.add_subplot(122)
plt.imshow(grad_y, cmap = 'gray')

plt.show()
```

```python
# Calculate an approximation of the gradient in that point by combining both results
sobel_edges = np.sqrt(grad_x ** 2 + grad_y ** 2)


plt.imshow(sobel_edges, cmap = 'gray')
plt.show()
```

```python
edges = cv2.Canny(I,threshold1=100,threshold2=255)

plt.imshow(edges, cmap = 'gray')
plt.show()
```

```python
g_edges = cv2.bitwise_and(sobel_edges,sobel_edges, mask = edges)
plt.imshow(g_edges, cmap = 'gray')
plt.show()
```
